```python
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
%matplotlib inline
```

```python
acc_valid = []
err_valid = []
err_map_valid = []
x_axis_valid = []
x_axis_main_train = []
x_axis_sub_train = []
acc_train = []
err_train = []
err_map_train = []
labels = ['joint_cost']  

plot_dir = '/Users/goncalocorreia/vqa_human_attention/expt/tuning/'
plot = 'joint_cost_second_att_plot_details.npz'

with np.load(plot_dir+plot) as data:
    err_valid.append(data['valid_error'])
    acc_valid.append(data['valid_accuracy'])
    x_axis_valid.append(data['x_axis_epochs'])
    err_map_valid.append(data['valid_error_map'])
    err_map_train.append(data['train_error_map'])
    x_axis_sub_train.append(data['sub_x_axis'])
    x_axis_main_train.append(data['main_x_axis'])
    err_train.append(data['train_error'])
    acc_train.append(data['train_accuracy'])
    
  
```

```python
fig = plt.figure(figsize=(12, 7))
ax1 = fig.add_subplot(1, 1, 1)
for i in xrange(len(x_axis_valid)):
    ax1.plot(x_axis_valid[i], err_map_valid[i], label='Val '+labels[i])
    ax1.plot(x_axis_sub_train[i], err_map_train[i], label='Train '+labels[i])
ax1.legend(loc=0,prop={'size':16})
ax1.set_xlabel('Epoch number')
ax1.set_ylabel('Human Att Map Error')
fig.show()
```

```python
fig = plt.figure(figsize=(12, 7))
ax1 = fig.add_subplot(1, 1, 1)
for i in xrange(len(x_axis_valid)):
    ax1.plot(x_axis_valid[i], acc_valid[i], label='Val '+labels[i])
    ax1.plot(x_axis_main_train[i], acc_train[i], label='Train '+labels[i])
ax1.legend(loc=0,prop={'size':16})
ax1.set_xlabel('Epoch number')
ax1.set_ylabel('Accuracy')
fig.show()
```

```python
fig = plt.figure(figsize=(12, 7))
ax1 = fig.add_subplot(1, 1, 1)
for i in xrange(len(x_axis_valid)):
    ax1.plot(x_axis_valid[i], err_valid[i], label='Val '+labels[i])
    ax1.plot(x_axis_main_train[i], err_train[i], label='Train '+labels[i])
ax1.legend(loc=0,prop={'size':16})
ax1.set_xlabel('Epoch number')
ax1.set_ylabel('Answer Error')
fig.show()
```

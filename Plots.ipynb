```python
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
%matplotlib inline
```

```python
err_map_valid = []
x_axis_valid = []
x_axis_sub_train = []
err_map_train = []
labels = ['first_att', 'second_att', 'no_learn']  
    
path = '/Users/goncalocorreia/vqa_human_attention/expt/train_att_maps/train_att_maps_first_att_orig_plot_details.npz'

with np.load(path) as data:
    x_axis_valid.append(data['x_axis_epochs'])
    err_map_valid.append(data['valid_error_map'])
    err_map_train.append(data['train_error_map'])
    x_axis_sub_train.append(data['sub_x_axis'])  
    
path = '/Users/goncalocorreia/vqa_human_attention/expt/train_att_maps/train_att_maps_reverse_kl_more_layers_drop_plot_details.npz'

with np.load(path) as data:
    x_axis_valid.append(data['x_axis_epochs'])
    err_map_valid.append(data['valid_error_map'])
    err_map_train.append(data['train_error_map'])
    x_axis_sub_train.append(data['sub_x_axis'])
    
path = '/Users/goncalocorreia/vqa_human_attention/expt/train_att_maps/train_att_maps_first_att_no_learn_plot_details.npz'

with np.load(path) as data:
    x_axis_valid.append(data['x_axis_epochs'])
    err_map_valid.append(data['valid_error_map'])
    err_map_train.append(data['train_error_map'])
    x_axis_sub_train.append(data['sub_x_axis'])  
```

```python
fig = plt.figure(figsize=(12, 7))
ax1 = fig.add_subplot(1, 1, 1)
for i in xrange(len(x_axis_valid)):
    ax1.plot(x_axis_valid[i], err_map_valid[i], label='Val '+labels[i])
    ax1.plot(x_axis_sub_train[i], err_map_train[i], label='Train '+labels[i])
ax1.legend(loc=0,prop={'size':16})
ax1.set_xlabel('Epoch number')
ax1.set_ylabel('Human Att Map Error')
fig.show()
```
